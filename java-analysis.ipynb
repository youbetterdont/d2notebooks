```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
cs_damage = lambda x: (x*28 - 278 + 1)/2.0
slvl = np.arange(28,60)
synergy_hard_points = 80
synergy_bonus = synergy_hard_points*0.1

synergized_dmg = cs_damage(slvl)*(1+synergy_bonus)

plt.figure()
plt.plot(slvl[1:], (synergized_dmg[1:]-synergized_dmg[:-1])/synergized_dmg[:-1]*100)
plt.xlabel('skill level')
plt.ylabel('percent dmg increase per skill point')
```
